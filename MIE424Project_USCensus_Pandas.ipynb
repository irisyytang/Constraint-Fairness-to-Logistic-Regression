Clean data
1. remove undesired attribute
2. remove instance with missing/incorrect arttribute value
3. random select 2w instances

```python
df1.describe().round(decimals=1)
```

```python
df1.columns
```

```python
# "i" means using original value, "d" means use mapped value e.g. continous to categorical
# df_clean = df[["dAge", "iCitizen","iDisabl1","iDisabl2",
#                     "iEnglish","iImmigr","iLang1",
#                     "iMarital","dOccup","iRelat1", "iRelat2",
#                     "iRlabor","dRpincome",
#                     "iSchool","iSex","iYearsch"]]

df_clean = df1[["dAge", "iCitizen","iDisabl1","iDisabl2",
                    "iEnglish","iImmigr","iLang1",
                    "iMarital","iRlabor", "iSchool", "iSex", "iYearsch", "iClass"]]
```

```python
df_clean
```

```python
df_clean['iClass_B'] = np.where( (df_clean['iClass'] == 0) | (df_clean['iClass'] == 1 ) | (df_clean['iClass'] == 2 ) | (df_clean['iClass'] == 6 ) | (df_clean['iClass'] == 7) | (df_clean['iClass'] == 8) | (df_clean['iClass'] == 9 ),0,1)
```

```python
df_clean1 = df_clean.loc[df_clean['iClass_B']== 1][:10000]
df_clean2 = df_clean.loc[df_clean['iClass_B']== 0][:10000]
df_clean_2w = df_clean1.append(df_clean2)
```

```python
df_clean_2w.describe().round(decimals=1)
```

```python
df_clean_2w.iClass_B.describe()
```

3. random select 2w instances

```python
# df_clean_2w = df_clean.sample(n=20000)
```

```python
df_clean_2w = df_clean_2w.sample(frac=1)
```

```python
df_clean_2w
```

```python
# df_clean_2w.to_csv()
compression_opts = dict(method='zip',
                        archive_name='USCensus1990_clean_iClass.csv')  
df_clean_2w.to_csv('USCensus1990_clean.zip', index=False,
          compression=compression_opts)  
```
